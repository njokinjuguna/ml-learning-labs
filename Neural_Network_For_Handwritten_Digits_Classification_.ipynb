```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train[0].shape
```

```python
X_train[0]
```

```python
plt.matshow(X_train[0])
```

flatten our data to prepare it for input layer of our neural network for handwritten digits classification

```python
X_train.shape
```

Lets scale our data for better accuracy results

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train_flattened=X_train.reshape(len(X_train),28*28)
X_train_flattened
```

```python
X_train_flattened.reshape(len(X_train),28*28)
X_train_flattened.shape
#we expect the flattening to be 6000,784
```

```python
X_test.reshape(len(X_test),28*28)
```

```python
X_test_flattened=X_test.reshape(len(X_test),28*28)
```

```python
X_test_flattened.shape
```

implementing the neural network

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
model.fit(X_train_flattened,y_train,epochs=5)
```

now lets evaluate how it will perform on the test data

```python
model.evaluate(X_test_flattened,y_test)
```

lets predict our model

```python
plt.matshow(X_test[0])
#
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted[0]
```

```python
np.argmax(y_predicted[0])
```

to see better our prediction lets apply the confusion matrix

```python
y_predicted
#but for us to run prediction  in confusion matrix we need y_predicted in the concreate form/whole number form
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
y_test[:5]
```

```python
confusion_matrix=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
confusion_matrix
```

lets try visualize the confusion matrix better in a graphical form

```python
import seaborn as sn
plt.figure(figsize=(10,10))
sn.heatmap(confusion_matrix,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

lets add a hidden layer to our neuro network to improve the accuracy

```python
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),  #hidden layer
    keras.layers.Dense(10,activation='sigmoid'),
])
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
model.fit(X_train_flattened,y_train,epochs=5)
```

lets evaluate the hidden layer

```python
model.evaluate(X_test_flattened,y_test)
```

lets see the confusion matrix after adding the hidden layer

```python
confusion_matrix=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
import seaborn as sn
plt.figure(figsize=(10,10))
sn.heatmap(confusion_matrix,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

lets make some adjustemnts ...instead of manually flattening the inputs,lets use the keras library flattening kit

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),  #hidden layer
    keras.layers.Dense(10,activation='sigmoid'),
])
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
model.fit(X_train,y_train,epochs=5)
```
