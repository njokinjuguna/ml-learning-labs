```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

## lets now flatten our dataset because as at now it is in 2d matrix of 28 by 28

```python
X_train.shape
```

```python
#lets scale our loaded data so we can get accurate results
X_train=X_train/255
X_test=X_test/255
```

```python
X_train
```

```python
#remember we need to convert(reshape) the 2D of 28 by 28...to flatten it,therefore we will multiple 28*28=784,then on the first parameter of reshape is the total rows.

X_train_flattened=X_train.reshape(len(X_train),28*28)
X_train_flattened
```

```python
X_test_flattened=X_test.reshape(len(X_test),28*28)
X_test_flattened
```

```python
X_train_flattened.shape
```

```python
X_test_flattened.shape
```

# Now lets create the neural network

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
model.fit(X_train_flattened,y_train,epochs=5)
```

# lets evaluate the model

```python
model.evaluate(X_test_flattened,y_test)
```

# lets predict the model

```python
plt.matshow(X_test[0])
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted[0]
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
y_test[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

# lets add a hidden layer to the model that we created

```python
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'), #the hidden layer added
    keras.layers.Dense(10,activation='sigmoid')
])
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
model.fit(X_train_flattened,y_train,epochs=5)
```

# evaluation of the new model with a hidden layer addition

```python
model.evaluate(X_test_flattened,y_test)
```

# prediction of the new model with a hidden layer addition

```python
y_predicted=model.predict(X_test_flattened)
y_predicted_labels=[np.argmax(i) for i in y_predicted]
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Prediction with the addition of the hidden layer')
plt.ylabel('Truth')
```

# finally ...remember we did flatten our array so we could train our model.however keras has a better way of doing it for us as shown below

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'), #the hidden layer added
    keras.layers.Dense(10,activation='sigmoid')
])
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
model.fit(X_train,y_train,epochs=5)
```
